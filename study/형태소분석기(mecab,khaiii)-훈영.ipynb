```python
review = '안녕하세요 꽃피우는시간입니다.실내반려식물케어 브랜드 꽃피우는시간이 온라인 런칭하며, 이용해주신 분들과 소통하기 위해 밴드를 개설했습니다.꽃피우는시간 제품 이용하시며 생긴 궁금증이나 문의사항을 공유해주시면 최선을 다해 답변드리겠습니다.감사합니다'

from konlpy.tag import Mecab
tagger = Mecab()

nouns = []
for noun in tagger.nouns(review):
        nouns.append(noun)
nouns[:10]
```

```python
#불용어 사전 만들기
stop_words = ['런']

#불용어 제외 후 형태소 분석 실행
nouns = []
for noun in tagger.nouns(review):
    if noun not in stop_words:
        nouns.append(noun)
nouns[:10]
```

```python
#빈도수 측정
from collections import Counter
nouns_counter = Counter(nouns)
top_nouns = dict(nouns_counter.most_common(50))
top_nouns 
```

```python
import numpy as np 
y_pos = np.arange(len(top_nouns))

plt.figure(figsize=(12,12))
plt.barh(y_pos, top_nouns.values())
plt.title('Word Count')
plt.yticks(y_pos, top_nouns.keys())
plt.show
```

```python
from wordcloud import WordCloud
wc = WordCloud(background_color='white', font_path='AppleGothic')
wc.generate_from_frequencies(top_nouns)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Mecab Nouns 전처리 전후비교

```python
diff_rows = 테스트[테스트['content_text_mecab'] != 테스트['original_text_mecab']]
diff_rows = diff_rows[['content_text', 'original_text', 'content_text_mecab', 'original_text_mecab']]
def find_unique_elements(df, column1, column2):
    unique_elements = []
    for i in range(len(df)):
        set1 = set(df[column1].iloc[i])
        set2 = set(df[column2].iloc[i])
        added_elements = list(set2 - set1)
        unique_elements.append(added_elements)
    return unique_elements

diff_rows['전처리 후 추가된 것'] = find_unique_elements(diff_rows, 'content_text_mecab', 'original_text_mecab')
diff_rows['전처리 후 삭제된 것'] = find_unique_elements(diff_rows, 'original_text_mecab', 'content_text_mecab')
diff_rows.head(3)
```

## 2. Mecab 모든 디갤 데이터에 워드 클라우드 적용해보기

```python
#title, content_text, original_text 각각 열에 형태소 분석기를 사용해볼 예정
테스트 = pd.read_csv('/Users/hollyyoon/Downloads/tree_compare.csv')
테스트.head(3)
```

```python
from konlpy.tag import Mecab
tagger = Mecab()

def mecab_noun(review):
    if isinstance(review, str):  # review 값이 문자열인 경우에만 처리
        nouns = []
        for noun in tagger.nouns(review):
                nouns.append(noun)
        return nouns
    else:
        return []

테스트['title_mecab'] = 테스트['title'].map(mecab_noun) 
테스트['content_text_mecab'] = 테스트['content_text'].map(mecab_noun) 
테스트['original_text_mecab'] = 테스트['original_text'].map(mecab_noun) 
테스트.head(3)
```

```python
def make_wordcloud_barplot(df, column_name, max):
    nouns = [item for sublist in df[column_name] for item in sublist]
    nouns_counter = Counter(nouns)
    top_nouns = dict(nouns_counter.most_common(max))

    y_pos = np.arange(len(top_nouns))

    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))
    
    ax1.barh(y_pos, top_nouns.values())
    ax1.set_title('Word Count')
    ax1.set_yticks(y_pos)
    ax1.set_yticklabels(top_nouns.keys())
    
    wc = WordCloud(background_color='white', font_path='AppleGothic')
    wc.generate_from_frequencies(top_nouns)

    ax2.imshow(wc, interpolation='bilinear')
    ax2.axis('off')

    plt.show()
```

```python
make_wordcloud_barplot(테스트, 'title_mecab', 30)
```

```python
make_wordcloud_barplot(테스트, 'original_text_mecab', 30)
```

```python
make_wordcloud_barplot(테스트, 'content_text_mecab', 30)
```

## Khaiii 테스트

```python
from khaiii import KhaiiiApi

# Khaiii 객체 생성
khaiii_api = KhaiiiApi()

# 분석할 텍스트
text = '안녕하세요 꽃피우는시간입니다.실내반려식물케어 브랜드 꽃피우는시간이 온라인 런칭하며, 이용해주신 분들과 소통하기 위해 밴드를 개설했습니다.꽃피우는시간 제품 이용하시며 생긴 궁금증이나 문의사항을 공유해주시면 최선을 다해 답변드리겠습니다.감사합니다'

# 형태소 분석
result = []
for word in khaiii_api.analyze(text):
    for morph in word.morphs:
        result.append((morph.lex, morph.tag))

# 결과 출력
print(result)
```

```python
from khaiii import KhaiiiApi, KhaiiiExcept
khaiii_api = KhaiiiApi()

def kha_keyword(review):
    if isinstance(review, str):  # review 값이 문자열인 경우에만 처리
        words = []
        try:
            for word in khaiii_api.analyze(review):
                for morph in word.morphs:
                    if morph.tag == 'NNG' and len(morph.lex) >= 2:
                        words.append(morph.lex)
        except KhaiiiExcept:
            pass
        return words
    else:
        return []
```

```python
테스트['title_kha'] = 테스트['title'].map(kha_keyword) 
테스트['content_text_kha'] = 테스트['content_text'].map(kha_keyword) 
테스트['original_text_kha'] = 테스트['original_text'].map(kha_keyword) 
테스트.head(3)
```

## Khaiii 사용자 사전 

```python
from khaiii import KhaiiiApi
```

```python
khaiii_api = KhaiiiApi()
stop_words = ['ㅋㅋ', '정도', '이상', '다음', '모습', '오늘']

def kha_keyword(review):
    if isinstance(review, str):  # review 값이 문자열인 경우에만 처리
        words = []
        try:
            for word in khaiii_api.analyze(review):
                for morph in word.morphs:
                    if morph.tag == 'NNG' and len(morph.lex) >= 2 and morph.lex not in stop_words:
                        words.append(morph.lex)
        except KhaiiiExcept:
            pass
        return words
    else:
        return []
    
테스트['title_kha'] = 테스트['title'].map(kha_keyword) 
테스트['content_text_kha'] = 테스트['content_text'].map(kha_keyword) 
테스트['original_text_kha'] = 테스트['original_text'].map(kha_keyword) 
테스트.head(3)
```
